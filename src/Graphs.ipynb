```python
def find_mode(ys):
    mode = max(ys)
    mode_index = ys.index(mode)
    return mode_index
```

```python
# u - index of mode
u1 = find_mode(y1)
u2 = find_mode(y2)
u3 = find_mode(y3)
u4 = find_mode(y4)

# t - index of item that is 2x of x of mode
t1 = x1.index(x1[u1] * 2)
t2 = x2.index(x2[u2] * 2)
t3 = x3.index(x3[u3] * 2)
t4 = x4.index(x4[u4] * 2)

xm1 = x1[t1:]
xm2 = x2[t2:]
xm3 = x3[t3:]
xm4 = x4[t4:]

ym1 = y1[t1:]
ym2 = y2[t2:]
ym3 = y3[t3:]
ym4 = y4[t4:]
```

```python
# N = n - 1, always even
def mean_time_fixed(N):
    sum = 0
    for k in range(0, N // 2):
        arg = (2 * k + 1) * np.pi / (2 * N)
        sum += ((-1) ** k) * alpha_fixed(N, arg) * (np.cos(arg) / (np.sin(arg) ** 3))
    sum *= 2 / N
    return sum

def alpha_fixed(N, arg):
    val = 1 - 1 / np.cosh((N / 2)  * (np.arccosh(1 + 2 * np.sin(arg) ** 2)))
    return val

def mean_time_og(N):
    sum = 0
    for k in range(0, N // 2):
        arg = (2 * k + 1) * np.pi / (2 * N)
        sum += ((-1) ** k) * alpha_og(N, arg) * (np.cos(arg) / (np.sin(arg) ** 3))
    sum *= 2 / N
    return sum

def alpha_og(N, arg):
    val = 1 - 1 / np.cosh(N / 2) / (np.cosh(1 + 2 * np.sin(arg) ** 2))
    return val


def get_2nd_term(i, j, k, N):
    term = 0
    for l in range(1, N):
        if (l % 2 == 1):
            term += ((math.sin(i * l * math.pi / N) * (math.tan(l * math.pi / (2 * N)) ** -1)) / 
                    ((math.sin(k * math.pi / (2 * N)) ** 2) + (math.sin(l * math.pi / (2 * N)) ** 2)))
    return term

def get_a_mx(N):
    a = []
    for i in range(0, N + 1):
        row = []
        for j in range(0, N + 1):
            val_ij = 0
            for k in range(1, N):
                if (k % 2 == 1):
                    val_ij += ((4 / (N ** 2)) * math.sin(j * k * math.pi / N) 
                               * (math.tan(k * math.pi / (2 * N)) ** -1) * get_2nd_term(i, j, k, N))
            row.append(val_ij)
        a.append(row)
    return a
```

```python
powers = np.arange(start=1, stop=5, step=1)
ns = 2 ** powers
func_vals_fixed = []
func_vals_og = []
func_vals_good = []
for N in ns:
    func_vals_fixed.append(mean_time_fixed(N))
    func_vals_og.append(mean_time_og(N))
    func_vals_good.append(get_a_mx(N)[N // 2][N // 2])
```

```python
plt.subplots(1, 1, dpi=200, figsize=(10, 5))
plt.plot(ns, func_vals_fixed, 'b', label='fixed')
plt.plot(ns, func_vals_og, 'r', label='og')
plt.plot(ns, func_vals_good, 'g', label='good')
plt.title("    ")
plt.xticks(ns)
plt.yticks(list(set(func_vals_fixed) | set(func_vals_og) | set(func_vals_good)))
plt.grid(True)
plt.legend()
```
