```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter, MaxNLocator

file_path = "D:\RW\\RWAnalyzer\\output\\eps_border\\bvp\\try_1\\absorption_time.txt"
with open(file_path, "r") as file:
    durations = [float(value) for value in file.readline().split()]

epsilon_values = np.arange(0.01, 0.71, 0.01)

plt.figure(figsize=(10, 6))
plt.scatter(epsilon_values, durations, color='blue')
#plt.plot(epsilon_values, durations, color='blue')

plt.xlabel('Предвзятость')
plt.ylabel('Количество очков')
plt.title('Зависимость количества очков от предвзятости, зависящей от расстояния до ближайшей границы,\nв случае игры человека за границу против компьютера')
plt.grid(True)
plt.legend()

plt.xticks(np.arange(0.0, 0.71, 0.1))

plt.ylim(durations[0] - 25, max(durations) + 25)
#plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True))
plt.yticks(list(plt.yticks()[0]) + [durations[0]])

output_file_path = "D:\RW\\RWAnalyzer\\output\\eps_border\\bvp\\try_1\\epsilon_vs_duration.png"
plt.savefig(output_file_path)

plt.show()
```
