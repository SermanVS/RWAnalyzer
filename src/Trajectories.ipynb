```python
games = []
cases = ["PvP", "PvE_center", "PvE_border", "BvB_sampling",
         "PvP_sampling", "PvP400+_sampling", "PvE_center_sampling", "PvE_border_sampling"]
lengths = ["short", "regular", "long"]

for case in cases:
    make_directories(case)
```

```python
# 7 here means that we consider 7 games out of all.

for case in cases:
    for length in lengths:
        for i in range (0, 7):
            directory = "../Trajectories/" + str(case) + "/" + length + "/"
            filename = case + "_" + length + "_" + str(i) + ".csv"
            try:
                game = pd.read_csv(directory + filename).values.tolist()
            except FileNotFoundError:
                pass
            games.append((case, length, game))
```

```python
i = 0
for game in games:
    trajectorise(game[2], game[0], game[1], i % 7)
    animate(game[0], game[1], i % 7)
    i += 1
```
